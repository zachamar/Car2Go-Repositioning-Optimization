## Section I: Introduction

$\quad$ American Office Systems, Inc was established in 1939. They manufacture small office equipment, overhead projectors and bookkeeping machines. Recently, they developed two versions of an adapter (one for IBM, one for Mac) that links a standard copy machine to personal computers. 
    
$\quad$ However, due to its small size and rather lackluster record, they were only able to secure 15 million as financing. They managed to build small facilities in 1994 and in 1995 the production and sales began. They now want to expand their production and have contacted us to help create an optimal solution to their problem: how to maximize the company's profits, given all the constraints they have.

## Section II: Initial Model 

### Assumptions

Our model relies on the following assumptions:

1. American Office Systems, Inc. runs 2 work shifts of 8 hours every day for 20 days per month. Each quarter is made of 3 months therefore we made our full capacity the number of adapters for IBM and MAC produced each hour multiplied by the hours worked in a quarter. 
2. Forecasted sales are accurate 
3. We immediately pay the interest for borrowing and capital in the previous quarter at the end of the following quarter 
4. Using a linear model is an appropriate approximation for an optimal strategy and further refinement will be done later. 
5. This model optimizes the accounting profit instead of the present and future cash flows of the operations. We opted for that sort of model because it depicts better the reality of the company and we found that, with the cash flow based models, the ending inventory of the last quarter was often too low to prepare for next year.
6. The fixed costs of 3 million dollars are for the whole company and only a portion is attributable to the adapters. We decided that 500,000 would be a good starting point.

### Defining Decision Variables

+ **CF** = Cash flow at the end of the quarter in units of dollars, in quarters i, where i = 1,2,3,4

### Step 4: Objective Function

$\quad$ Our objective is to maximize the profits of American Office Systems, Inc. To do so we will maximize what will bring revenue and minimize the costs. This allows us to develop the following objective function:

$$\scriptsize\max\sum_{i\in\{0,1,2,3\}}\bigg((Price_{IBM}*S_{i_i}) + (Price_{MAC}*S_{m_i}) - (VC_{ibm}*S_{i_i}) - (VC_{mac}*S_{m_i}) - CostOfAdvertising*(Mi{i_i} + Mm{i_i})) - 0.03*(VC_{Mac}*Em{i_i}+VC_{Ibm}*Ei{i_i}) \bigg)
    \\ \scriptsize+ CF_{0}*(((1+InterestRateBonds)^{0.25})-1))
    \\ \scriptsize+ CF_{1}*(((1+InterestRateBonds)^{0.25})-1))
    \\ \scriptsize+ CF_{2}*(((1+InterestRateBonds)^{0.25})-1))
    \\ \scriptsize- B_{0}*((1+InterestRateLOC)^{0.25}-1)
    \\ \scriptsize- B_{1}*((1+InterestRateLOC)^{0.25}-1) 
    \\ \scriptsize- B_{2}*((1+InterestRateLOC)^{0.25}-1)
    \\ \scriptsize- 4*FixedCostsPerQuarter
    \\ \scriptsize- 4*DepreciationPerQuarter$$
    
$\quad$ In words, what this function does is: start with the revenue, both IBM and MAC and add the interest rate on returns from the bond. We did not include the last quarter's bond income as it will affect next years first quarter not this year. To that we subtract the variable costs, the fixed costs, the depreciation, the advertising costs, the carrying costs and the loan interest (only for 3 quarters as the first quarter's loan is paid back in the next).

### Key findings from the optimal solution: 


**See Appendix 1: Optimal Solution**

$\quad$ This model tried to maximize the profit for American Office Systems while respecting the several constraints that came up during the difficult time it was going through. In this case, the optimal solution resulted in a profit of almost 9M ($ 8,981,990.88) at the end of the 4th quarter. 

$\quad$ In terms of production, the optimal solution suggested producing as much as we could all the time. From Q1 to Q3  we produced 28 800 for IBM and 9 600 for Mac. in Q4 however, with the building being finished we could produce an extra 10% of each, leading to 31 680 IBM adapters and 10 500 Mac. This will allow the compnay to be more prepared for the extra demande the fourth quarter has (probably due to holidat seasons) and will prepare them for next year.

$\quad$ Since we can and have to produce more in that fourth quarter, we make use of the line of credit. We borrow the full amount (750K). By looking closely at the ending inventory in the last quarter we see that it fulfilled its purpose since it is 4 times bigger than the one in Q3 for both types of adapters. Thus, respecting the constraint requiring extra inventory for the begining of the following year.

$\quad$ The company started the year strong, making the as many sales as possible sales in Q1 for both adapters. Whe looking at the numbers it is shocking to see how much is spent in advertising. In fact during Q1, 77% of IBM and 87% of MAC sales were due to marketing efforts. For the rest of the year, the sales were a lot lower than in Q1, they remained constant through the middle quarters with a final considerable decrease in Q4. 

$\quad$ The trend is similar for the marketing effort. In Q2 and Q3 it also lowered compared to Q1. It went down to  around 60% for IBM and 77% for Mac. Finally, in Q4 it dropped again representing 45% of the total sales for IBM and 67% of Mac. We could assume that since the company had to spend more in production for that quarter, there were less funds available for the advertising of the units, resulting in less sales overall as it did not seem to be a capacity constraint since the company ended with extra inventory.

$\quad$ We can conclude from this optimal solution that American Office Systems will have a busy year producing at maximum capacity and advertising a lot. They will however be rewarded with 9M$ of profits which can go a long way for them. This would already be more than half of what they had obainned as financing.

```python
#initilizing empty dictionaries
Pi = {}  
Pm = {} 
Si = {} 
Sm = {} 
Mi = {} 
Mm = {} 
B = {} 
Ei = {}
Em = {} 
CF = {}

# Create decisions variable
for i in np.arange(4):  
    Pi[i] = model.addVar(lb=0)
    Pm[i] = model.addVar(lb=0)
    Si[i] = model.addVar(lb=0)
    Sm[i] = model.addVar(lb=0)
    Mi[i] = model.addVar(lb=0)
    Mm[i] = model.addVar(lb=0)
    B[i] = model.addVar(lb=0)
    CF[i] = model.addVar(lb=0)
    Ei[i] = model.addVar(lb=0)
    Em[i] = model.addVar(lb=0)
    # It is optional to specify the lower and upper bounds of the variable, as the defaults are suitable for our case
```

#### Step 4: Constrainst

```python
# Production ibm constraint
con_production_ibm = {}
con_production_ibm[0] = model.addConstr(Pi[0]<=28800)
con_production_ibm[1] = model.addConstr(Pi[1]<=28800)
con_production_ibm[2] = model.addConstr(Pi[2]<=28800)
con_production_ibm[3] = model.addConstr(Pi[3]<=31680)

# Production mac constraint
con_production_mac = {}
con_production_mac[0] = model.addConstr(Pm[0]<=9600)
con_production_mac[1] = model.addConstr(Pm[1]<=9600)
con_production_mac[2] = model.addConstr(Pm[2]<=9600)
con_production_mac[3] = model.addConstr(Pm[3]<=10560)

# Inventory management constraint ibm
estimated_sale_ibm = {}
estimated_sale_ibm[0] = model.addConstr(Pi[0] + Ei[0] >= 1.1*(10000+Mi[1]))
estimated_sale_ibm[1] = model.addConstr(Pi[1] + Ei[1] >= 1.1*(11000+Mi[2]))
estimated_sale_ibm[2] = model.addConstr(Pi[2] + Ei[2] >= 1.1*(12000+Mi[3]))
estimated_sale_ibm[3] = model.addConstr(Pi[3] + Ei[3] >= 2*(12000+Mi[3]))

# Inventory management constraint mac
estimated_sale_mac = {}
estimated_sale_mac[0] = model.addConstr(Pm[0] + Em[0] >= 1.1*(2000+Mm[1]))
estimated_sale_mac[1] = model.addConstr(Pm[1] + Em[1] >= 1.1*(2200+Mm[2]))
estimated_sale_mac[2] = model.addConstr(Pm[2] + Em[2] >= 1.1*(2400+Mm[3]))
estimated_sale_mac[3] = model.addConstr(Pm[3] + Em[3] >= 2*(2400+Mm[3]))

# Sales Contraints IBM
sales_IBM = {}
sales_IBM[0] = model.addConstr(Si[0] <= 9000 + Mi[0])
sales_IBM[1] = model.addConstr(Si[1] <= 10000 + Mi[1])
sales_IBM[2] = model.addConstr(Si[2] <= 11000 + Mi[2])
sales_IBM[3] = model.addConstr(Si[3] <= 12000 + Mi[3])

# Sales Contraints Mac
sales_MAC = {}
sales_MAC[0] = model.addConstr(Sm[0] <= 1800 + Mm[0])
sales_MAC[1] = model.addConstr(Sm[1] <= 2000 + Mm[1])
sales_MAC[2] = model.addConstr(Sm[2] <= 2200 + Mm[2])
sales_MAC[3] = model.addConstr(Sm[3] <= 2400 + Mm[3])

# Ending inventory ibm 
ending_inventory_ibm = {}
ending_inventory_ibm[0] = model.addConstr(Ei[0] == beginning_inventory_ibm + Pi[0] - Si[0])
ending_inventory_ibm[1] = model.addConstr(Ei[1] == Ei[0] + Pi[1] - Si[1])
ending_inventory_ibm[2] = model.addConstr(Ei[2] == Ei[1] + Pi[2] - Si[2])
ending_inventory_ibm[3] = model.addConstr(Ei[3] == Ei[2] + Pi[3] - Si[3])

# Ending inventory mac
ending_inventory_mac = {}
ending_inventory_mac[0] = model.addConstr(Em[0] == beginning_inventory_mac + Pm[0] - Sm[0])
ending_inventory_mac[1] = model.addConstr(Em[1] == Em[0] + Pm[1] - Sm[1])
ending_inventory_mac[2] = model.addConstr(Em[2] == Em[1] + Pm[2] - Sm[2])
ending_inventory_mac[3] = model.addConstr(Em[3] == Em[2] + Pm[3] - Sm[3])

# Maximum Credit Line
credit_limit = {}
credit_limit[0] = model.addConstr(B[0] <= 750000)
credit_limit[1] = model.addConstr(B[1] <= 750000)
credit_limit[2] = model.addConstr(B[2] <= 750000)
credit_limit[3] = model.addConstr(B[3] <= 750000)

# Cash flow
cash_flow = {}

cash_flow[0] = model.addConstr(CF[0] == (B[0] 
                                        + price_ibm*(0.75*Si[0] + 0.20*8000 + 0.03*7000 + 0.02*7000*0.5) 
                                        + price_mac*(0.80*Sm[0] + 0.11*1600 + 0.05*1400 + 0.04*1400*0.5) 
                                        - (VC_ibm*Pi[0] + cost_advertising*Mi[0]) 
                                        - (VC_mac*Pm[0] + cost_advertising*Mm[0])
                                        - fixed_costs 
                                        - 0.03*(VC_mac*Em[0]+VC_ibm*Ei[0])))


cash_flow[1] = model.addConstr(CF[1] == ((B[1]-B[0])
                                       + price_ibm*(0.75*Si[1] + 0.20*Si[0] + 0.03*8000 + 0.02*8000*0.5) 
                                       + price_mac*(0.80*Sm[1] + 0.11*Sm[0] + 0.05*1600 + 0.04*1600*0.5) 
                                       - (VC_ibm*Pi[1] + cost_advertising*Mi[1]) 
                                       - (VC_mac*Pm[1] + cost_advertising*Mm[1]) 
                                       - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
                                       - fixed_costs 
                                       - 0.03*(VC_mac*Em[1]+VC_ibm*Ei[1])
                                       +((1+interest_rate_bonds)**(0.25))*CF[0]))


cash_flow[2] = model.addConstr(CF[2] == ((B[2]-B[1]) 
                                       + price_ibm*(0.75*Si[2] + 0.20*Si[1] + 0.03*Si[0] + 0.02*Si[0]*0.5) 
                                       + price_mac*(0.80*Sm[2] + 0.11*Sm[1] + 0.05*Sm[0] + 0.04*Sm[0]*0.5) 
                                       - (VC_ibm*Pi[2] + cost_advertising*Mi[2]) 
                                       - (VC_mac*Pm[2] + cost_advertising*Mm[2]) 
                                       - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
                                       - fixed_costs
                                       - 0.03*(VC_mac*Em[2]+VC_ibm*Ei[2])
                                       + ((1+interest_rate_bonds)**(0.25))*CF[1]))


cash_flow[3] = model.addConstr(CF[3] == ((B[3]-B[2]) 
                                       + price_ibm*(0.75*Si[3] + 0.20*Si[2] + 0.03*Si[1] + 0.02*Si[1]*0.5) 
                                       + price_mac*(0.80*Sm[3] + 0.11*Sm[2] + 0.05*Sm[1] + 0.04*Sm[1]*0.5) 
                                       - (VC_ibm*Pi[3] + cost_advertising*Mi[3])
                                       - (VC_mac*Pm[3] + cost_advertising*Mm[3]) 
                                       - B[2]*((1+interest_rate_LOC)**(0.25)-1) 
                                       - fixed_costs
                                       - 0.03*(VC_mac*Em[3]+VC_ibm*Ei[3])
                                       + ((1+interest_rate_bonds)**(0.25))*CF[2]))
```

#### Step 5: Objective Function

```python
# objective function
obj = ( # + Revenue
          price_ibm*Si[0] + price_mac*Sm[0]
        + price_ibm*Si[1] + price_mac*Sm[1]
        + price_ibm*Si[2] + price_mac*Sm[2]
        + price_ibm*Si[3] + price_mac*Sm[3]
    
        # + Interest on returns - We did not include last term as it will affect next year
        + ((1+interest_rate_bonds)**(0.25)-1)*CF[0]
        + ((1+interest_rate_bonds)**(0.25)-1)*CF[1]
        + ((1+interest_rate_bonds)**(0.25)-1)*CF[2]
            
        # - Variable Costs
        - VC_ibm*Si[0] - VC_mac*Sm[0] 
        - VC_ibm*Si[1] - VC_mac*Sm[1]
        - VC_ibm*Si[2] - VC_mac*Sm[2]
        - VC_ibm*Si[3] - VC_mac*Sm[3]
        
        # - Fixed Costs
        - 4*fixed_costs

        # - Depreciation
        - 4*depreciation
    
        # - Advertising Costs
        - cost_advertising*(Mi[0]+Mi[1]+Mi[2]+Mi[3] + Mm[0]+Mm[1]+Mm[2]+Mm[3])
        
        # - Carrying Costs
        - 0.03*(VC_mac*Em[0]+VC_ibm*Ei[0])
        - 0.03*(VC_mac*Em[1]+VC_ibm*Ei[1])
        - 0.03*(VC_mac*Em[2]+VC_ibm*Ei[2])
        - 0.03*(VC_mac*Em[3]+VC_ibm*Ei[3])
        
        # - Loan Interest
        - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
        - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
        - B[2]*((1+interest_rate_LOC)**(0.25)-1)
            
      )    
        
model.setObjective(obj, GRB.MAXIMIZE)
```

```python
model.optimize()
```

```python
# Define the write off arrays for IBM and MAC before the sensitivity and cost it will have

def sensitivity_AC(IBM, MAC, cost):
    
    model = Model('case study')  # object
    
    #initilizing empty dictionaries
    Pi = {}  
    Pm = {} 
    Si = {} 
    Sm = {} 
    Mi = {} 
    Mm = {} 
    B = {} 
    Ei = {}
    Em = {} 
    CF = {}

    # Create decisions variable
    for i in np.arange(4):  
        Pi[i] = model.addVar(lb=0)
        Pm[i] = model.addVar(lb=0)
        Si[i] = model.addVar(lb=0)
        Sm[i] = model.addVar(lb=0)
        Mi[i] = model.addVar(lb=0)
        Mm[i] = model.addVar(lb=0)
        B[i] = model.addVar(lb=0)
        CF[i] = model.addVar(lb=0)
        Ei[i] = model.addVar(lb=0)
        Em[i] = model.addVar(lb=0)
    
        # Constraints
    # Production ibm constraint
    con_production_ibm = {}
    con_production_ibm[0] = model.addConstr(Pi[0]<=28800)
    con_production_ibm[1] = model.addConstr(Pi[1]<=28800)
    con_production_ibm[2] = model.addConstr(Pi[2]<=28800)
    con_production_ibm[3] = model.addConstr(Pi[3]<=31680)

    # Production mac constraint
    con_production_mac = {}
    con_production_mac[0] = model.addConstr(Pm[0]<=9600)
    con_production_mac[1] = model.addConstr(Pm[1]<=9600)
    con_production_mac[2] = model.addConstr(Pm[2]<=9600)
    con_production_mac[3] = model.addConstr(Pm[3]<=10560)

    # Inventory management constraint ibm
    estimated_sale_ibm = {}
    estimated_sale_ibm[0] = model.addConstr(Pi[0] + Ei[0] >= 1.1*(10000+Mi[1]))
    estimated_sale_ibm[1] = model.addConstr(Pi[1] + Ei[1] >= 1.1*(11000+Mi[2]))
    estimated_sale_ibm[2] = model.addConstr(Pi[2] + Ei[2] >= 1.1*(12000+Mi[3]))
    estimated_sale_ibm[3] = model.addConstr(Pi[3] + Ei[3] >= 2*(12000+Mi[3]))

    # Inventory management constraint mac
    estimated_sale_mac = {}
    estimated_sale_mac[0] = model.addConstr(Pm[0] + Em[0] >= 1.1*(2000+Mm[1]))
    estimated_sale_mac[1] = model.addConstr(Pm[1] + Em[1] >= 1.1*(2200+Mm[2]))
    estimated_sale_mac[2] = model.addConstr(Pm[2] + Em[2] >= 1.1*(2400+Mm[3]))
    estimated_sale_mac[3] = model.addConstr(Pm[3] + Em[3] >= 2*(2400+Mm[3]))

    # Sales Contraints IBM
    sales_IBM = {}
    sales_IBM[0] = model.addConstr(Si[0] <= 9000 + Mi[0])
    sales_IBM[1] = model.addConstr(Si[1] <= 10000 + Mi[1])
    sales_IBM[2] = model.addConstr(Si[2] <= 11000 + Mi[2])
    sales_IBM[3] = model.addConstr(Si[3] <= 12000 + Mi[3])

    # Sales Contraints Mac
    sales_MAC = {}
    sales_MAC[0] = model.addConstr(Sm[0] <= 1800 + Mm[0])
    sales_MAC[1] = model.addConstr(Sm[1] <= 2000 + Mm[1])
    sales_MAC[2] = model.addConstr(Sm[2] <= 2200 + Mm[2])
    sales_MAC[3] = model.addConstr(Sm[3] <= 2400 + Mm[3])

    # Ending inventory ibm 
    ending_inventory_ibm = {}
    ending_inventory_ibm[0] = model.addConstr(Ei[0] == beginning_inventory_ibm + Pi[0] - Si[0])
    ending_inventory_ibm[1] = model.addConstr(Ei[1] == Ei[0] + Pi[1] - Si[1])
    ending_inventory_ibm[2] = model.addConstr(Ei[2] == Ei[1] + Pi[2] - Si[2])
    ending_inventory_ibm[3] = model.addConstr(Ei[3] == Ei[2] + Pi[3] - Si[3])

    # Ending inventory mac
    ending_inventory_mac = {}
    ending_inventory_mac[0] = model.addConstr(Em[0] == beginning_inventory_mac + Pm[0] - Sm[0])
    ending_inventory_mac[1] = model.addConstr(Em[1] == Em[0] + Pm[1] - Sm[1])
    ending_inventory_mac[2] = model.addConstr(Em[2] == Em[1] + Pm[2] - Sm[2])
    ending_inventory_mac[3] = model.addConstr(Em[3] == Em[2] + Pm[3] - Sm[3])

    # Maximum Credit Line
    credit_limit = {}
    credit_limit[0] = model.addConstr(B[0] <= 750000)
    credit_limit[1] = model.addConstr(B[1] <= 750000)
    credit_limit[2] = model.addConstr(B[2] <= 750000)
    credit_limit[3] = model.addConstr(B[3] <= 750000)

    # Cash flow
    cash_flow = {}
    
    cash_flow[0] = model.addConstr(CF[0] == (B[0] 
                                           + price_ibm*(IBM[0]*Si[0] + IBM[1]*8000 + IBM[2]*7000 + IBM[3]*7000*0.5) 
                                           + price_mac*(MAC[0]*Sm[0] + MAC[1]*1600 + MAC[2]*1400 + MAC[3]*1400*0.5) 
                                           - (VC_ibm*Pi[0] + cost_advertising*Mi[0]) 
                                           - (VC_mac*Pm[0] + cost_advertising*Mm[0])
                                           - fixed_costs - cost/4
                                           - 0.03*(VC_mac*Em[0]+VC_ibm*Ei[0])))

    cash_flow[1] = model.addConstr(CF[1] == ((B[1]-B[0])
                                           + price_ibm*(IBM[0]*Si[1] + IBM[1]*Si[0] + IBM[2]*8000 + IBM[3]*8000*0.5) 
                                           + price_mac*(MAC[0]*Sm[1] + MAC[1]*Sm[0] + MAC[2]*1600 + MAC[3]*1600*0.5) 
                                           - (VC_ibm*Pi[1] + cost_advertising*Mi[1]) 
                                           - (VC_mac*Pm[1] + cost_advertising*Mm[1]) 
                                           - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs - cost/4
                                           - 0.03*(VC_mac*Em[1]+VC_ibm*Ei[1])
                                           +((1+interest_rate_bonds)**(0.25))*CF[0]))


    cash_flow[2] = model.addConstr(CF[2] == ((B[2]-B[1]) 
                                           + price_ibm*(IBM[0]*Si[2] + IBM[1]*Si[1] + IBM[2]*Si[0] + IBM[3]*Si[0]*0.5) 
                                           + price_mac*(MAC[0]*Sm[2] + MAC[1]*Sm[1] + MAC[2]*Sm[0] + MAC[3]*Sm[0]*0.5) 
                                           - (VC_ibm*Pi[2] + cost_advertising*Mi[2]) 
                                           - (VC_mac*Pm[2] + cost_advertising*Mm[2]) 
                                           - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs - cost/4
                                           - 0.03*(VC_mac*Em[2]+VC_ibm*Ei[2])
                                           + ((1+interest_rate_bonds)**(0.25))*CF[1]))


    cash_flow[3] = model.addConstr(CF[3] == ((B[3]-B[2]) 
                                           + price_ibm*(IBM[0]*Si[3] + IBM[1]*Si[2] + IBM[2]*Si[1] + IBM[3]*Si[1]*0.5) 
                                           + price_mac*(MAC[0]*Sm[3] + MAC[1]*Sm[2] + MAC[2]*Sm[1] + MAC[3]*Sm[1]*0.5) 
                                           - (VC_ibm*Pi[3] + cost_advertising*Mi[3])
                                           - (VC_mac*Pm[3] + cost_advertising*Mm[3]) 
                                           - B[2]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs - cost/4
                                           - 0.03*(VC_mac*Em[3]+VC_ibm*Ei[3])
                                           + ((1+interest_rate_bonds)**(0.25))*CF[2]))

    
    #Objective Function
    
    obj = ( # + Revenue
          price_ibm*Si[0] + price_mac*Sm[0]
        + price_ibm*Si[1] + price_mac*Sm[1]
        + price_ibm*Si[2] + price_mac*Sm[2]
        + price_ibm*Si[3] + price_mac*Sm[3]
    
        # + Interest on returns - We did not include last term as it will affect next year
        + ((1+interest_rate_bonds)**(0.25)-1)*CF[0]
        + ((1+interest_rate_bonds)**(0.25)-1)*CF[1]
        + ((1+interest_rate_bonds)**(0.25)-1)*CF[2]
            
        # - Variable Costs
        - VC_ibm*Si[0] - VC_mac*Sm[0] 
        - VC_ibm*Si[1] - VC_mac*Sm[1]
        - VC_ibm*Si[2] - VC_mac*Sm[2]
        - VC_ibm*Si[3] - VC_mac*Sm[3]
        
        # - Fixed Costs (and cost of project)
        - 4*fixed_costs - cost

        # - Depreciation
        - 4*depreciation
    
        # - Advertising Costs
        - cost_advertising*(Mi[0]+Mi[1]+Mi[2]+Mi[3] + Mm[0]+Mm[1]+Mm[2]+Mm[3])
        
        # - Carrying Costs
        - 0.03*(VC_mac*Em[0]+VC_ibm*Ei[0])
        - 0.03*(VC_mac*Em[1]+VC_ibm*Ei[1])
        - 0.03*(VC_mac*Em[2]+VC_ibm*Ei[2])
        - 0.03*(VC_mac*Em[3]+VC_ibm*Ei[3])
        
        # - Loan Interest
        - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
        - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
        - B[2]*((1+interest_rate_LOC)**(0.25)-1)
            
      )       

    model.setObjective(obj, GRB.MAXIMIZE)        
    
    model.optimize()
    
    return model.objVal
    
```

```python
def sensitivity_production(rate_increase):
    model = Model('case study')  # object
    
    #initilizing empty dictionaries
    Pi = {}  
    Pm = {} 
    Si = {} 
    Sm = {} 
    Mi = {} 
    Mm = {} 
    B = {} 
    Ei = {}
    Em = {} 
    CF = {}

    # Create decisions variable
    for i in np.arange(4):  
        Pi[i] = model.addVar(lb=0)
        Pm[i] = model.addVar(lb=0)
        Si[i] = model.addVar(lb=0)
        Sm[i] = model.addVar(lb=0)
        Mi[i] = model.addVar(lb=0)
        Mm[i] = model.addVar(lb=0)
        B[i] = model.addVar(lb=0)
        CF[i] = model.addVar(lb=0)
        Ei[i] = model.addVar(lb=0)
        Em[i] = model.addVar(lb=0)
    
        # Constraints
    # Production ibm constraint
    con_production_ibm = {}
    con_production_ibm[0] = model.addConstr(Pi[0]<=28800)
    con_production_ibm[1] = model.addConstr(Pi[1]<=28800)
    con_production_ibm[2] = model.addConstr(Pi[2]<=28800)
    con_production_ibm[3] = model.addConstr(Pi[3]<=28800*(1+rate_increase))

    # Production mac constraint
    con_production_mac = {}
    con_production_mac[0] = model.addConstr(Pm[0]<=9600)
    con_production_mac[1] = model.addConstr(Pm[1]<=9600)
    con_production_mac[2] = model.addConstr(Pm[2]<=9600)
    con_production_mac[3] = model.addConstr(Pm[3]<=9600*(1+rate_increase))

    # Inventory management constraint ibm
    estimated_sale_ibm = {}
    estimated_sale_ibm[0] = model.addConstr(Pi[0] + Ei[0] >= 1.1*(10000+Mi[1]))
    estimated_sale_ibm[1] = model.addConstr(Pi[1] + Ei[1] >= 1.1*(11000+Mi[2]))
    estimated_sale_ibm[2] = model.addConstr(Pi[2] + Ei[2] >= 1.1*(12000+Mi[3]))
    estimated_sale_ibm[3] = model.addConstr(Pi[3] + Ei[3] >= 2*(12000+Mi[3]))

    # Inventory management constraint mac
    estimated_sale_mac = {}
    estimated_sale_mac[0] = model.addConstr(Pm[0] + Em[0] >= 1.1*(2000+Mm[1]))
    estimated_sale_mac[1] = model.addConstr(Pm[1] + Em[1] >= 1.1*(2200+Mm[2]))
    estimated_sale_mac[2] = model.addConstr(Pm[2] + Em[2] >= 1.1*(2400+Mm[3]))
    estimated_sale_mac[3] = model.addConstr(Pm[3] + Em[3] >= 2*(2400+Mm[3]))

    # Sales Contraints IBM
    sales_IBM = {}
    sales_IBM[0] = model.addConstr(Si[0] <= 9000 + Mi[0])
    sales_IBM[1] = model.addConstr(Si[1] <= 10000 + Mi[1])
    sales_IBM[2] = model.addConstr(Si[2] <= 11000 + Mi[2])
    sales_IBM[3] = model.addConstr(Si[3] <= 12000 + Mi[3])

    # Sales Contraints Mac
    sales_MAC = {}
    sales_MAC[0] = model.addConstr(Sm[0] <= 1800 + Mm[0])
    sales_MAC[1] = model.addConstr(Sm[1] <= 2000 + Mm[1])
    sales_MAC[2] = model.addConstr(Sm[2] <= 2200 + Mm[2])
    sales_MAC[3] = model.addConstr(Sm[3] <= 2400 + Mm[3])

    # Ending inventory ibm 
    ending_inventory_ibm = {}
    ending_inventory_ibm[0] = model.addConstr(Ei[0] == beginning_inventory_ibm + Pi[0] - Si[0])
    ending_inventory_ibm[1] = model.addConstr(Ei[1] == Ei[0] + Pi[1] - Si[1])
    ending_inventory_ibm[2] = model.addConstr(Ei[2] == Ei[1] + Pi[2] - Si[2])
    ending_inventory_ibm[3] = model.addConstr(Ei[3] == Ei[2] + Pi[3] - Si[3])

    # Ending inventory mac
    ending_inventory_mac = {}
    ending_inventory_mac[0] = model.addConstr(Em[0] == beginning_inventory_mac + Pm[0] - Sm[0])
    ending_inventory_mac[1] = model.addConstr(Em[1] == Em[0] + Pm[1] - Sm[1])
    ending_inventory_mac[2] = model.addConstr(Em[2] == Em[1] + Pm[2] - Sm[2])
    ending_inventory_mac[3] = model.addConstr(Em[3] == Em[2] + Pm[3] - Sm[3])

    # Maximum Credit Line
    credit_limit = {}
    credit_limit[0] = model.addConstr(B[0] <= 750000)
    credit_limit[1] = model.addConstr(B[1] <= 750000)
    credit_limit[2] = model.addConstr(B[2] <= 750000)
    credit_limit[3] = model.addConstr(B[3] <= 750000)

    # Cash flow
    cash_flow = {}

    cash_flow[0] = model.addConstr(CF[0] == (B[0] 
                                            + price_ibm*(0.75*Si[0] + 0.20*8000 + 0.03*7000 + 0.02*7000*0.5) 
                                            + price_mac*(0.80*Sm[0] + 0.11*1600 + 0.05*1400 + 0.04*1400*0.5) 
                                            - (VC_ibm*Pi[0] + cost_advertising*Mi[0]) 
                                            - (VC_mac*Pm[0] + cost_advertising*Mm[0])
                                            - fixed_costs 
                                            - 0.03*(VC_mac*Em[0]+VC_ibm*Ei[0])))


    cash_flow[1] = model.addConstr(CF[1] == ((B[1]-B[0])
                                           + price_ibm*(0.75*Si[1] + 0.20*Si[0] + 0.03*8000 + 0.02*8000*0.5) 
                                           + price_mac*(0.80*Sm[1] + 0.11*Sm[0] + 0.05*1600 + 0.04*1600*0.5) 
                                           - (VC_ibm*Pi[1] + cost_advertising*Mi[1]) 
                                           - (VC_mac*Pm[1] + cost_advertising*Mm[1]) 
                                           - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs 
                                           - 0.03*(VC_mac*Em[1]+VC_ibm*Ei[1])
                                           +((1+interest_rate_bonds)**(0.25))*CF[0]))


    cash_flow[2] = model.addConstr(CF[2] == ((B[2]-B[1]) 
                                           + price_ibm*(0.75*Si[2] + 0.20*Si[1] + 0.03*Si[0] + 0.02*Si[0]*0.5) 
                                           + price_mac*(0.80*Sm[2] + 0.11*Sm[1] + 0.05*Sm[0] + 0.04*Sm[0]*0.5) 
                                           - (VC_ibm*Pi[2] + cost_advertising*Mi[2]) 
                                           - (VC_mac*Pm[2] + cost_advertising*Mm[2]) 
                                           - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs
                                           - 0.03*(VC_mac*Em[2]+VC_ibm*Ei[2])
                                           + ((1+interest_rate_bonds)**(0.25))*CF[1]))


    cash_flow[3] = model.addConstr(CF[3] == ((B[3]-B[2]) 
                                           + price_ibm*(0.75*Si[3] + 0.20*Si[2] + 0.03*Si[1] + 0.02*Si[1]*0.5) 
                                           + price_mac*(0.80*Sm[3] + 0.11*Sm[2] + 0.05*Sm[1] + 0.04*Sm[1]*0.5) 
                                           - (VC_ibm*Pi[3] + cost_advertising*Mi[3])
                                           - (VC_mac*Pm[3] + cost_advertising*Mm[3]) 
                                           - B[2]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs
                                           - 0.03*(VC_mac*Em[3]+VC_ibm*Ei[3])
                                           + ((1+interest_rate_bonds)**(0.25))*CF[2]))

    # objective function
    obj = ( # + Revenue
              price_ibm*Si[0] + price_mac*Sm[0]
            + price_ibm*Si[1] + price_mac*Sm[1]
            + price_ibm*Si[2] + price_mac*Sm[2]
            + price_ibm*Si[3] + price_mac*Sm[3]

            # + Interest on returns - We did not include last term as it will affect next year
            + ((1+interest_rate_bonds)**(0.25)-1)*CF[0]
            + ((1+interest_rate_bonds)**(0.25)-1)*CF[1]
            + ((1+interest_rate_bonds)**(0.25)-1)*CF[2]

            # - Variable Costs
            - VC_ibm*Si[0] - VC_mac*Sm[0] 
            - VC_ibm*Si[1] - VC_mac*Sm[1]
            - VC_ibm*Si[2] - VC_mac*Sm[2]
            - VC_ibm*Si[3] - VC_mac*Sm[3]

            # - Fixed Costs
            - 4*fixed_costs

            # - Depreciation
            - 4*depreciation

            # - Advertising Costs
            - cost_advertising*(Mi[0]+Mi[1]+Mi[2]+Mi[3] + Mm[0]+Mm[1]+Mm[2]+Mm[3])

            # - Carrying Costs
            - 0.03*(VC_mac*Em[0]+VC_ibm*Ei[0])
            - 0.03*(VC_mac*Em[1]+VC_ibm*Ei[1])
            - 0.03*(VC_mac*Em[2]+VC_ibm*Ei[2])
            - 0.03*(VC_mac*Em[3]+VC_ibm*Ei[3])

            # - Loan Interest
            - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
            - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
            - B[2]*((1+interest_rate_LOC)**(0.25)-1)

          ) 
    model.setObjective(obj, GRB.MAXIMIZE)        
    
    model.optimize()
    
    return model.objVal
```

```python
def sensitivity_VC(ibm_VC_change, mac_VC_change):
    
    model = Model('case study')  # object
    
    #initilizing empty dictionaries
    Pi = {}  
    Pm = {} 
    Si = {} 
    Sm = {} 
    Mi = {} 
    Mm = {} 
    B = {} 
    Ei = {}
    Em = {} 
    CF = {}

    # Create decisions variable
    for i in np.arange(4):  
        Pi[i] = model.addVar(lb=0)
        Pm[i] = model.addVar(lb=0)
        Si[i] = model.addVar(lb=0)
        Sm[i] = model.addVar(lb=0)
        Mi[i] = model.addVar(lb=0)
        Mm[i] = model.addVar(lb=0)
        B[i] = model.addVar(lb=0)
        CF[i] = model.addVar(lb=0)
        Ei[i] = model.addVar(lb=0)
        Em[i] = model.addVar(lb=0)      
     
        # Constraints
    # Production ibm constraint
    con_production_ibm = {}
    con_production_ibm[0] = model.addConstr(Pi[0]<=28800)
    con_production_ibm[1] = model.addConstr(Pi[1]<=28800)
    con_production_ibm[2] = model.addConstr(Pi[2]<=28800)
    con_production_ibm[3] = model.addConstr(Pi[3]<=31680)

    # Production mac constraint
    con_production_mac = {}
    con_production_mac[0] = model.addConstr(Pm[0]<=9600)
    con_production_mac[1] = model.addConstr(Pm[1]<=9600)
    con_production_mac[2] = model.addConstr(Pm[2]<=9600)
    con_production_mac[3] = model.addConstr(Pm[3]<=10560)

    # Inventory management constraint ibm
    estimated_sale_ibm = {}
    estimated_sale_ibm[0] = model.addConstr(Pi[0] + Ei[0] >= 1.1*(10000+Mi[1]))
    estimated_sale_ibm[1] = model.addConstr(Pi[1] + Ei[1] >= 1.1*(11000+Mi[2]))
    estimated_sale_ibm[2] = model.addConstr(Pi[2] + Ei[2] >= 1.1*(12000+Mi[3]))
    estimated_sale_ibm[3] = model.addConstr(Pi[3] + Ei[3] >= 2*(12000+Mi[3]))

    # Inventory management constraint mac
    estimated_sale_mac = {}
    estimated_sale_mac[0] = model.addConstr(Pm[0] + Em[0] >= 1.1*(2000+Mm[1]))
    estimated_sale_mac[1] = model.addConstr(Pm[1] + Em[1] >= 1.1*(2200+Mm[2]))
    estimated_sale_mac[2] = model.addConstr(Pm[2] + Em[2] >= 1.1*(2400+Mm[3]))
    estimated_sale_mac[3] = model.addConstr(Pm[3] + Em[3] >= 2*(2400+Mm[3]))

    # Sales Contraints IBM
    sales_IBM = {}
    sales_IBM[0] = model.addConstr(Si[0] <= 9000 + Mi[0])
    sales_IBM[1] = model.addConstr(Si[1] <= 10000 + Mi[1])
    sales_IBM[2] = model.addConstr(Si[2] <= 11000 + Mi[2])
    sales_IBM[3] = model.addConstr(Si[3] <= 12000 + Mi[3])

    # Sales Contraints Mac
    sales_MAC = {}
    sales_MAC[0] = model.addConstr(Sm[0] <= 1800 + Mm[0])
    sales_MAC[1] = model.addConstr(Sm[1] <= 2000 + Mm[1])
    sales_MAC[2] = model.addConstr(Sm[2] <= 2200 + Mm[2])
    sales_MAC[3] = model.addConstr(Sm[3] <= 2400 + Mm[3])

    # Ending inventory ibm 
    ending_inventory_ibm = {}
    ending_inventory_ibm[0] = model.addConstr(Ei[0] == beginning_inventory_ibm + Pi[0] - Si[0])
    ending_inventory_ibm[1] = model.addConstr(Ei[1] == Ei[0] + Pi[1] - Si[1])
    ending_inventory_ibm[2] = model.addConstr(Ei[2] == Ei[1] + Pi[2] - Si[2])
    ending_inventory_ibm[3] = model.addConstr(Ei[3] == Ei[2] + Pi[3] - Si[3])

    # Ending inventory mac
    ending_inventory_mac = {}
    ending_inventory_mac[0] = model.addConstr(Em[0] == beginning_inventory_mac + Pm[0] - Sm[0])
    ending_inventory_mac[1] = model.addConstr(Em[1] == Em[0] + Pm[1] - Sm[1])
    ending_inventory_mac[2] = model.addConstr(Em[2] == Em[1] + Pm[2] - Sm[2])
    ending_inventory_mac[3] = model.addConstr(Em[3] == Em[2] + Pm[3] - Sm[3])

    # Maximum Credit Line
    credit_limit = {}
    credit_limit[0] = model.addConstr(B[0] <= 750000)
    credit_limit[1] = model.addConstr(B[1] <= 750000)
    credit_limit[2] = model.addConstr(B[2] <= 750000)
    credit_limit[3] = model.addConstr(B[3] <= 750000)

    # Cash flow
    cash_flow = {}

    cash_flow[0] = model.addConstr(CF[0] == (B[0] 
                                            + price_ibm*(0.75*Si[0] + 0.20*8000 + 0.03*7000 + 0.02*7000*0.5) 
                                            + price_mac*(0.80*Sm[0] + 0.11*1600 + 0.05*1400 + 0.04*1400*0.5) 
                                            - ((VC_ibm+ibm_VC_change)*Pi[0] + cost_advertising*Mi[0]) 
                                            - ((VC_mac+mac_VC_change)*Pm[0] + cost_advertising*Mm[0]) 
                                            - fixed_costs 
                                            - 0.03*((VC_mac+mac_VC_change)*Em[0]+(VC_ibm+ibm_VC_change)*Ei[0])))


    cash_flow[1] = model.addConstr(CF[1] == ((B[1]-B[0])
                                           + price_ibm*(0.75*Si[1] + 0.20*Si[0] + 0.03*8000 + 0.02*8000*0.5) 
                                           + price_mac*(0.80*Sm[1] + 0.11*Sm[0] + 0.05*1600 + 0.04*1600*0.5) 
                                           - ((VC_ibm+ibm_VC_change)*Pi[1] + cost_advertising*Mi[1]) 
                                           - ((VC_mac+mac_VC_change)*Pm[1] + cost_advertising*Mm[1]) 
                                           - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs 
                                           - 0.03*((VC_mac+mac_VC_change)*Em[1]+(VC_ibm+ibm_VC_change)*Ei[1])
                                           +((1+interest_rate_bonds)**(0.25))*CF[0]))


    cash_flow[2] = model.addConstr(CF[2] == ((B[2]-B[1]) 
                                           + price_ibm*(0.75*Si[2] + 0.20*Si[1] + 0.03*Si[0] + 0.02*Si[0]*0.5) 
                                           + price_mac*(0.80*Sm[2] + 0.11*Sm[1] + 0.05*Sm[0] + 0.04*Sm[0]*0.5) 
                                           - ((VC_ibm+ibm_VC_change)*Pi[2] + cost_advertising*Mi[2]) 
                                           - ((VC_mac+mac_VC_change)*Pm[2] + cost_advertising*Mm[2]) 
                                           - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs
                                           - 0.03*((VC_mac+mac_VC_change)*Em[2]+(VC_ibm+ibm_VC_change)*Ei[2])
                                           + ((1+interest_rate_bonds)**(0.25))*CF[1]))


    cash_flow[3] = model.addConstr(CF[3] == ((B[3]-B[2]) 
                                           + price_ibm*(0.75*Si[3] + 0.20*Si[2] + 0.03*Si[1] + 0.02*Si[1]*0.5) 
                                           + price_mac*(0.80*Sm[3] + 0.11*Sm[2] + 0.05*Sm[1] + 0.04*Sm[1]*0.5) 
                                           - ((VC_ibm+ibm_VC_change)*Pi[3] + cost_advertising*Mi[3])
                                           - ((VC_ibm+ibm_VC_change)*Pm[3] + cost_advertising*Mm[3]) 
                                           - B[2]*((1+interest_rate_LOC)**(0.25)-1) 
                                           - fixed_costs
                                           - 0.03*((VC_mac+mac_VC_change)*Em[3]+(VC_ibm+ibm_VC_change)*Ei[3])
                                           + ((1+interest_rate_bonds)**(0.25))*CF[2]))

    # objective function
    obj = ( # + Revenue
              price_ibm*Si[0] + price_mac*Sm[0]
            + price_ibm*Si[1] + price_mac*Sm[1]
            + price_ibm*Si[2] + price_mac*Sm[2]
            + price_ibm*Si[3] + price_mac*Sm[3]

            # + Interest on returns - We did not include last term as it will affect next year
            + ((1+interest_rate_bonds)**(0.25)-1)*CF[0]
            + ((1+interest_rate_bonds)**(0.25)-1)*CF[1]
            + ((1+interest_rate_bonds)**(0.25)-1)*CF[2]

            # - Variable Costs
            - (VC_ibm+ibm_VC_change)*Si[0] - (VC_mac+mac_VC_change)*Sm[0] 
            - (VC_ibm+ibm_VC_change)*Si[1] - (VC_mac+mac_VC_change)*Sm[1]
            - (VC_ibm+ibm_VC_change)*Si[2] - (VC_mac+mac_VC_change)*Sm[2]
            - (VC_ibm+ibm_VC_change)*Si[3] - (VC_mac+mac_VC_change)*Sm[3]

            # - Fixed Costs
            - 4*fixed_costs

            # - Depreciation
            - 4*depreciation

            # - Advertising Costs
            - cost_advertising*(Mi[0]+Mi[1]+Mi[2]+Mi[3] + Mm[0]+Mm[1]+Mm[2]+Mm[3])

            # - Carrying Costs
            - 0.03*(VC_mac*Em[0]+VC_ibm*Ei[0])
            - 0.03*(VC_mac*Em[1]+VC_ibm*Ei[1])
            - 0.03*(VC_mac*Em[2]+VC_ibm*Ei[2])
            - 0.03*(VC_mac*Em[3]+VC_ibm*Ei[3])

            # - Loan Interest
            - B[0]*((1+interest_rate_LOC)**(0.25)-1) 
            - B[1]*((1+interest_rate_LOC)**(0.25)-1) 
            - B[2]*((1+interest_rate_LOC)**(0.25)-1)

          )    
        
    model.setObjective(obj, GRB.MAXIMIZE)        
    
    model.optimize()
    
    return model.objVal
```
